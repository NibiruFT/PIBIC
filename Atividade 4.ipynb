## Simule um canal binário simétrico para $\alpha=0,1$ e para $\alpha=0,001$ e mostre graficamente a probabilidade de erro da simulação $\hat \alpha$ em função do número de bits, $n=1,2,3,...,N$.

## Para $\alpha=0,1$

```python
import numpy as np
from scipy.stats import binom
from matplotlib import pyplot as plt
Calcule a precisão relativa da simulação:
		- $$\varepsilon=\frac{\sqrt{var(\hat\alpha)}}\alpha  $$
#----------------------------------------------------------------------
# This function adjusts matplotlib settings for a uniform feel in the textbook.
# Note that with usetex=True, fonts are rendered with LaTeX.  This may
# result in an error if LaTeX is not installed on your system.  In that case,
# you can set usetex to False.
if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=18, usetex=True)

#------------------------------------------------------------
# Define the distribution parameters to be plotted
n_values = [10, 100, 1000]
alpha_values = [0.1, 0.1, 0.1]
linestyles = ['-', '--', ':']
x = np.arange(-1, 20000)

#------------------------------------------------------------
# plot the distributions
fig, ax = plt.subplots(figsize=(10, 10))

for (n, alpha, ls) in zip(n_values, alpha_values, linestyles):
    # create a binomial distribution
    dist = binom(n, alpha)

    plt.plot(x, dist.pmf(x), color='blue', linestyle=ls,
             label=r'$\alpha=%.1f,\ n=%i$' % (alpha, n))

plt.xlim(-10, 150)
plt.ylim(0, 0.6)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\alpha, n)$')
plt.title('Distribuição Binomial')

plt.legend()
plt.show()
```

## Calcule a precisão relativa da simulação:

## $$\varepsilon=\frac{\sqrt{var(\hat\alpha)}}\alpha  $$


# Variância:

## Para $\alpha=0,1$:

### 	 A variância de uma distribuição de Bernoulli é definida por $Var(X)=E(X^2)-[E(X)]^2$. Temos que:
#### 		 $$E(X^2)=0^2\times P(X=0)+1^2\times P(X=1)=0\times(1-\alpha)+1\times \alpha=\alpha$$
### 	 Assim, 
### 		 $$Var(X)=\alpha-\alpha^2=\alpha(1-\alpha)$$ 
###    Substituindo, temos:
###        $$=0.1(1-0.1)=0.09 $$

# Precisão Relativa:

## $$\varepsilon=\frac{\sqrt{0.09}}{0.09}=3.33  $$

---
## Para $\alpha=0,001$:

###    Substituindo, temos:
###       $$=0.001(1-0.001)=0.000009 $$

# Precisão Relativa:

## $$\varepsilon=\frac{\sqrt{0.000009}}{0.000009}=333.33  $$
