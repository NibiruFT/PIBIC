# Atividade 4.1 - Precisão Relativa

### Para $\alpha$ = 0.1

```python
import random
import matplotlib.pyplot as plt
import numpy as np

def erro():
    
    alfa = 0.1
    
    m = round(random.random()<alfa)
    if(m==0):
        return('Acerto')
    else:
        return('Erro')
def repete(n):
    
    alfa = 0.1
    
    nacertos = nerros = 0
    facertos = []
    ferros = []
    for i in range(n):
        facertos.append(0) 
        ferros.append(0)
    for j in range(n):
       arremesso = erro()
       
       if(arremesso == 'Acerto'):
           nacertos += 1
       else:
           nerros += 1

        ### guarda as frequências relativas parciais 
       facertos[j] = nacertos/(j+1)
       ferros[j] = nerros/(j+1)

    ### Apresenta as frequências relativas finais 
    print('Precisão Relativa = ', np.sqrt(np.var(ferros))/alfa)
    
    ###  Mostra graficamente
    x = np.arange(1,n+1,1)
    plt.figure(figsize=(10,5),dpi=70)
    #plt.plot(x,facertos,label='acerto')
    plt.plot(x,ferros,label='erro')
    plt.title('Convergência para o valor real de alfa')
    plt.ylabel('Prob. de erro')
    plt.xlabel('n')
    plt.minorticks_on()
    plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
    plt.legend()    
    plt.show()
    
def menu():
    """
    Menu de entrada do número de bits
    
    """
    n = int(input('Quantos bits de entrada? '))
    repete(n)
    
    
while True:
    menu()
       
```
